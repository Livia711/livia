```python
import pandas as pd
df_selic = pd.read_json('https://api.bcb.gov.br/dados/serie/bcdata.sgs.11/dados?formato=json')

print(df_selic.info())
print(df_selic)
```

```python
# verificar duplicidade de linhas (PASSO MUITO IMPORTANTE), utilizando a função drop_duplicates()
df_selic.drop_duplicates(keep= 'last', inplace=True)
# mantem  o ultimo registro (keep='last')
# através do parâmetro INPLACE=TRUE, faz com que a transformação seja sava no Dataframe
# No nosso caso, nao esxite linha duplicada
```

```python
# adicionando colunas
from datetime import date
from datetime import datetime as dt

data_estracao = date.today()

df_selic['data_estracao'] = data_estracao
df_selic['responsavel'] = "Lívia"

print (df_selic.info)
df_selic.head(2)
```

```python
import pandas as pd
# Criando um DataFrame com 5 linhas de dados
data = {
    'nome': [ 'Produto A', 'Produto B', 'Produto C', 'Produto D', 'Produto E'],
    'quantidade de itens comprados': [3, 1, 4, 3, 2],
    'tipo de item': ['Eletrônico', 'Vestuario','Alimento', 'Eletrônico', 'Vestuario'], # added missing comma to separate list items
    'receita total': [120, 80, 60, 120, 90]
}
df = pd.DataFrame(data)
print(df)
```

```python
# Duplicando uma linha
df.drop_duplicates(keep= 'last', inplace=True)
print(df)
```

```python
# Calculando a coluna 'preço do intem'
df['preço do item'] = df['receita total'] / df['quantidade de itens comprados']

# Selecionando preço do intem acima de 50 reais
itens_acima_de_50 = df[df['preço do item'] > 50]
print(itens_acima_de_50)
```
